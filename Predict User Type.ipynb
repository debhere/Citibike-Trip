```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('citibike_tripdata.csv')
df.sample(15)
```

```python
df.info()
```

```python
df['name_localizedValue'].unique()
```

```python
df.loc[df['usertype'] == 'Customer']
```

```python
#Localizedvalue and usertype very much correlated, so an remove localizedvalue column
#Remove start/stop time, station id, bike id, start station name, stop name

tripdata = df.drop(columns = ['starttime', 'stoptime', 'start station id', 'start station name', 'end station id', 
                              'end station name', 'bikeid', 'name_localizedValue'], axis=1)

tripdata['usertype'] = tripdata['usertype'].apply(lambda x : 1 if x == 'Subscriber' else 0)
tripdata
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = tripdata[['usertype']]
X = tripdata[['tripduration']]
```

```python
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
acc = accuracy_score(y_val, y_pred)
acc
```

```python
pre = precision_score(y_val, y_pred)
pre
```

```python
recall = recall_score(y_val, y_pred)
recall
```

```python
f1 = f1_score(y_val, y_pred)
f1
```
